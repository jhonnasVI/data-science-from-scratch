Finding Key Connectors

```python
users = [
    {"id": 0, "name": "Hero"},
    {"id": 1, "name": "Dunn"},
    {"id": 2, "name": "Sue"},
    {"id": 3, "name": "Chi"},
    {"id": 4, "name": "Thor"},
    {"id": 5, "name": "Clive"},
    {"id": 6, "name": "Hicks"},
    {"id": 7, "name": "Devin"},
    {"id": 8, "name": "Kate"},
    {"id": 9, "name": "Klein"}
]
```

Friendship data represented as a list of pairs of IDs

```python
friendships = [
    (0, 1),
    (0, 2),
    (1, 2),
    (1, 3),
    (2, 3),
    (3, 4),
    (4, 5),
    (5, 6),
    (5, 7),
    (6, 8),
    (7, 8),
    (8, 9)
]
```

Set each user's friends property to an empty list

```python
for user in users:
    user["friends"] = []
```

Populate the lists using the friendship data

```python
for i, j in friendships:
    # users[i] is the user whose id is i
    users[i]["friends"].append(users[j]) # add i as friend of j
    users[j]["friends"].append(users[i]) # add j as friend of i
```

What is the average number of connections?

- find the total number of connections, by summing up the lengths of all the friends list

```python
def numberOfFriends(user):
    # how many friends does the user have?
    return len(user["friends"])

totalConnections = sum(numberOfFriends(user) for user in users)
print(totalConnections)
```

- divide by the number of users

```python
numberOfUsers = len(users)
averageConnection = totalConnections / numberOfUsers

print(averageConnection)
```

- sort from "most friends" to "least friends"

```python
numberOfFriendsById = [
    (user["id"], numberOfFriends(user)) for user in users
]
# get it sorted by numberOfFriends largest to smallest
sorted(numberOfFriendsById, key=lambda x: x[1], reverse=True)
```

This has been sorted from the largest connection to the smallest connection.

- id: 1 (Dunn) has 3 connections: (0: Hero, 2: Sue, 3: Chi)
- id: 2 (Sue) has 3 connections: (0: Hero, 1: Dunn, 3: Chi)
- id: 3 (Chi) has 3 connections: (2: Sue, 1: Dunn, 4: Thor)
- id: 5 (Clive) has 3 connections: (6: Hicks, 7: Devin, 4: Thor)
- id: 8 (Kate) has 3 connections: (6: Hicks, 7: Devin, 9: Klein)
- id: 0 (Hero) has 2 connections: (2: Sue, 1: Dunn)
- id: 4 (Thor) has 2 connections: (3: Chi, 5: Clive)
- id: 6 (Hicks) has 2 connections: (5: Clive, 8: Kate)
- id: 7 (Devin) has 2 connections: (5: Clive, 8: Kate)
- id: 9 (Klein) has 1 connection: (8: Kate)

This is the computation of network metric degree centrality.
