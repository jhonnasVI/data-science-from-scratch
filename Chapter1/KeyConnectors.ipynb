Finding Key Connectors

```python
users = [
    {"id": 0, "name": "Hero"},
    {"id": 1, "name": "Dunn"},
    {"id": 2, "name": "Sue"},
    {"id": 3, "name": "Chi"},
    {"id": 4, "name": "Thor"},
    {"id": 5, "name": "Clive"},
    {"id": 6, "name": "Hicks"},
    {"id": 7, "name": "Devin"},
    {"id": 8, "name": "Kate"},
    {"id": 9, "name": "Klein"}
]
```

Friendship data represented as a list of pairs of IDs

```python
friendship_pairs = [
    (0, 1),
    (0, 2),
    (1, 2),
    (1, 3),
    (2, 3),
    (3, 4),
    (4, 5),
    (5, 6),
    (5, 7),
    (6, 8),
    (7, 8),
    (8, 9)
]
```

Set each user's friends property to an empty list

```python
friendships = {
    user["id"]: [] for user in users
}
```

Populate the lists using the friendship data

```python
for i, j in friendship_pairs:
    friendships[i].append(j) # add j as friend of i
    friendships[j].append(i) # add i as friend of j
```

What is the average number of connections?

- find the total number of connections, by summing up the lengths of all the friends list

```python
def numberOfFriends(user):
    # how many friends does the user have?
    userId = user["id"]
    friendIds = friendships[userId]
    return len(friendIds)

totalConnections = sum(numberOfFriends(user) for user in users)
print(totalConnections)
```

- divide by the number of users

```python
numberOfUsers = len(users)
averageConnection = totalConnections / numberOfUsers

print(averageConnection)
```

- sort from "most friends" to "least friends"

```python
numberOfFriendsByid = [
    (user["id"], numberOfFriends(user)) for user in users
]

numberOfFriendsByid.sort(
    key=lambda idAndFriends: idAndFriends[1], reverse=True
)

print(numberOfFriendsByid)
```

This has been sorted from the largest connection to the smallest connection.

- id: 1 (Dunn) has 3 connections: (0: Hero, 2: Sue, 3: Chi)
- id: 2 (Sue) has 3 connections: (0: Hero, 1: Dunn, 3: Chi)
- id: 3 (Chi) has 3 connections: (2: Sue, 1: Dunn, 4: Thor)
- id: 5 (Clive) has 3 connections: (6: Hicks, 7: Devin, 4: Thor)
- id: 8 (Kate) has 3 connections: (6: Hicks, 7: Devin, 9: Klein)
- id: 0 (Hero) has 2 connections: (2: Sue, 1: Dunn)
- id: 4 (Thor) has 2 connections: (3: Chi, 5: Clive)
- id: 6 (Hicks) has 2 connections: (5: Clive, 8: Kate)
- id: 7 (Devin) has 2 connections: (5: Clive, 8: Kate)
- id: 9 (Klein) has 1 connection: (8: Kate)

This is the computation of network metric degree centrality.

Data Scientist You May Know

```python
def friendOfAFriendIds(user):
    return [
        foafId 
        for friendId in friendships[user["id"]]
        for foafId in friendships[friendId]
    ]

hero = friendOfAFriendIds(users[0])
print(hero)
```

- id: 0 (hero) appears twice, because he is friend with id: 2 (Sue) and id: 1 (Dunn). Also, id: 3 (Chi) appears twice, as Chi is friends with Sue and Dunn.

Producing a count of mutual friends, and exclude people already known to the user.

```python
from collections import Counter

def friendsOfFriends(user):
    userId = user["id"]

    return Counter(
        foafId
        # for each of my friends,
        for friendId in friendships[userId]
        # find their friends
        for foafId in friendships[friendId]
        # who's not me and not my friends.
        if foafId != userId and foafId not in friendships[userId]
    )

print(friendsOfFriends(users[3]))
```

This tells that user id: 3 (Chi):
- has 2 mutual friends with id: 0 (Hero)
- has 1 mutual friend with id: 5 (Clive)
